```python
#visualizing existing data
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(data=iris, hue="species", size=1.5)
```

```python
X_iris = iris.drop('species', axis=1)
X_iris.shape
```

```python
Y_iris = iris['species']
Y_iris.shape
```

```python
#Creating my own DataFrame
import pandas as pd
words_num_dict = {0:12, 1:9, 2:10, 3:1, 4:0, 5:2, 6:3}
italicness_dict = {0:0.01, 1:0., 2:0.95, 3:0., 4:0., 5:0.98, 6:0}
words_num = pd.Series(words_num_dict)
italicness = pd.Series(italicness_dict)
lines = pd.DataFrame({'words_num':words_num, 'italicness':italicness})
lines
```

```python
lines.index
```

```python
lines.columns
```

```python
#DataFrame from a list of dictionaries
dicts = [{'wn':12, 'ital':0}, {'wn':8, 'ital':.95}, {'wn':2, 'ital':.5}, {'wn':1, 'ital':0.}]
lines = pd.DataFrame(dicts)
lines
```

```python
#DataFrame from a 2d array
lines_arr = [[12, 0.], [2, .95], [2, 0.], [5, .99], [3, .75]]
lines = pd.DataFrame(lines_arr, columns = ['words_num', 'italic'])
lines
```

```python
# Trying Naive Bayes Classification on Iris
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, Y_iris,random_state=1)
from sklearn.naive_bayes import GaussianNB # 1. choose model class
model = GaussianNB()                       # 2. instantiate model
model.fit(Xtrain, ytrain)                  # 3. fit model to data
y_model = model.predict(Xtest)             # 4. predict on new data
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```
