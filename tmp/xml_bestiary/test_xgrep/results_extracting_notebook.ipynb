```python
import pandas as pd
import seaborn as sns
lines_vectors = pd.read_csv('../../../training_data.csv')
lines_vectors.head()
```

```python
X_lines = lines_vectors.drop('type', axis=1)
Y_lines = lines_vectors['type']
X_lines.head()
```

```python
sns.pairplot(data=lines_vectors, hue="type", height=6)
```

```python
# Trying Naive Bayes Classification on lines
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X_lines, Y_lines,random_state=1)
from sklearn.naive_bayes import GaussianNB # 1. choose model class
model = GaussianNB()                       # 2. instantiate model
model.fit(Xtrain, Ytrain)                  # 3. fit model to data
y_model = model.predict(Xtest)             # 4. predict on new data
from sklearn.metrics import accuracy_score
accuracy_score(Ytest, y_model)
```
